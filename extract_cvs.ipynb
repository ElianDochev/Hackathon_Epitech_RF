```python
import pandas as pd
df = pd.read_csv("data/transcripts/whisper.csv", nrows= 10)
df
```

```python
import json
import os

if not os.path.exists("results"):
    os.makedirs("results")
for index, dict_obj in enumerate(df["segments"]):
    with open(f"results/{index}.json", "w") as f:
        json.dump(dict_obj, f)
```

```python
# open the cvs files in data/grid and get the entry with the magnetothequeId that matches the one we get form df_two
import os
import pandas as pd

df_array = []

for root, dirs, files in os.walk("data/grid"):
    for file in files:
        if file.endswith(".csv"):
            df_new = pd.read_csv(f"data/grid/{file}")
            df_array.append(df_new)

df_two = pd.concat(df_array)
df_two
```

```python
import json
import os
import pandas as pd

if not os.path.exists("results4"):
    os.makedirs("results4")


for index, dict_obj in enumerate(df["segments"]):
    # create a file with the name results/{index}.json
    extracted_data = [obj["text"] for obj in json.loads(dict_obj)]
    with open(f"results4/trascpirt_{df['magnetothequeId'][index]}.txt", "w") as f:
        res = "".join(extracted_data)
        f.write("Transcpirt: \n")
        f.write(res)
        # find the entry in df_two that matches the magnetothequeId in df
        lookup_id = df["magnetothequeId"][index]
        lookup_thing = df_two[df_two["magnetotheque_id"] == lookup_id]
        # get diffusion_title, concept_title, diffusion_standfirst and collapse them into a string
        diffusion_title = lookup_thing["diffusion_title"].to_string(index=False)
        concept_title = lookup_thing["concept_title"].to_string(index=False)
        diffusion_standfirst = lookup_thing["diffusion_standfirst"].to_string(index=False)
        f.write("\n")
        f.write("diffusion_title: \n")
        f.write(diffusion_title)
        f.write("\n")
        f.write("concept_title: \n")
        f.write(concept_title)
        f.write("\n")
        f.write("diffusion_standfirst: \n")
        f.write(diffusion_standfirst)
```
